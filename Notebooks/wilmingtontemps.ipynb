```python
# Importing libraries
import hvplot.pandas
import pandas as pd

import warnings

warnings.filterwarnings('ignore')
```

```python
# Define the URL for NC mean temperature data download

nc_temp_url = (
    'https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/city/time'
    '-series/USW00013748/tmax/12/12/1943-2023.csv')
```

```python
# Import NC temperature values from NCEI

nc_temp_df = pd.read_csv(
    nc_temp_url, header=3, names=['year', 'temperature_f'])
```

```python
# Extract the year from the date

nc_temp_df.year = pd.to_datetime(nc_temp_df.year, format='%Y%m').dt.year

```

```python
# Compute celsius temperatures

nc_temp_df['temperature_c'] = (nc_temp_df['temperature_f'] - 32) * 5 / 9
```

```python
# Plot the NC temperature data over time

nc_temp_df.plot(
    x='year', y='temperature_f',
    title='Maximum Annual Temperature in Wilmington, NC, USA',
    xlabel='Year', ylabel='Temperature(F)');
```
